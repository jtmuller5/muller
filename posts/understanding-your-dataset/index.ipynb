Almost every time I start a new project, I begin by performing a cursory analysis of the data set. This sort of high-level analysis gives me a general idea about the kind of data I'm working with, the size of the data set, and the types of relationships I may be interested in exploring.

I will be using the [UFO Sightings](https://www.kaggle.com/datasets/willianoliveiragibin/ufo-sightings/data) dataset in this post to demonstrate useful methods for understanding your data. To get started, open the dataset in Kaggle and press the "Download" button and move the `ufo-sightings-transformed.csv` file into your working directory.

Install `ipykernel` if you haven't already so you can keep your Jupyter notebook environment isolated.

````julia
import pandas as pd

file_path = 'ufo-sightings-transformed.csv'
df = pd.read_csv(file_path)
````

1. Display the First Few Rows

````julia
df.head()
````

2. Display the Last Few Rows

````julia
df.tail()
````

3. Display Basic Information

````julia
df.info()
````

4. Display Summary Statistics

````julia
# Display summary statistics for numerical columns
df.describe()
````

5. Display Column Names

````julia
# Display the column names
df.columns
````

6. Display Data Types of Columns

````julia
# Display data types of each column
df.dtypes
````

7. Check for Missing Values

````julia
# Check for missing values in each column
df.isnull().sum()
````

8. Display Unique Values in a Column

````julia
# Display unique values in a specific column (replace 'column_name' with the actual column name)
df['Season'].unique()
````

9. Display Value Counts for a Column

````julia
# Display value counts for a specific column (replace 'column_name' with the actual column name)
df['UFO_shape'].value_counts()
````

10. Display the Shape of the DataFrame

````julia
# Display the shape of the DataFrame (number of rows and columns)
df.shape
````

11. Display Correlation Matrix

````julia
# Display the correlation matrix for numerical columns
df.corr(numeric_only=True)
````

12. Plot Histograms for Numerical Columns

````julia
%pip install matplotlib
````

````julia
import matplotlib.pyplot as plt

# Plot histograms for numerical columns
df.hist(figsize=(10, 10), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
````

13. Plot a Heatmap of Missing Values (using [seaborn](https://pypi.org/project/seaborn/))

````julia
%pip install seaborn
````

````julia
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a heatmap of missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
````

14. Plot Pairwise Relationships

````julia
# Plot pairwise relationships in the dataset (useful for numerical columns)
# Warning - this may take a long time to execute if the dataset is large
sns.pairplot(df)
plt.show()
````

15. Display a Random Sample of Rows

````julia
# Display a random sample of rows (replace 5 with the desired number of rows)
df.sample(5)
````
